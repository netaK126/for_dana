```python
import numpy as np
import torch
```

```python
train = "social_honeypot.train.csv"
test = "social_honeypot.test.csv"

X_train = np.loadtxt(train, delimiter=',', usecols=list(range(1, 15+1)))
y_train = np.loadtxt(train, delimiter=',', usecols=0).astype(np.int32)
#X_train = X_train[:, top_idx]
print(X_train.shape)
X_test = np.loadtxt(test, delimiter=',', usecols=list(range(1, 15+1)))
y_test = np.loadtxt(test, delimiter=',', usecols=0).astype(np.int32)
#X_test = X_test[:, top_idx]
print(X_test.shape)
```

```python
X_combined = np.concatenate((X_train, X_test), axis=0)
X_combined_mean = np.mean(X_combined,axis=0)
X_combined_std = np.std(X_combined,axis=0)
print(X_combined_mean)
print(X_combined_std)




print((X_combined).shape)
#X_train = (X_train-X_combined_mean)/X_combined_std
#X_test = (X_test-X_combined_mean)/X_combined_std

X_train = (X_train)/np.max(X_combined,axis=0)
X_test = (X_test)/np.max(X_combined,axis=0)


train_data = torch.from_numpy(X_train.astype(np.float32))
train_labels = torch.from_numpy(y_train.astype(np.float32))

test_data = torch.from_numpy(X_test.astype(np.float32))
test_labels = torch.from_numpy(y_test.astype(np.float32))

print(train_data, train_data.size())
print(train_labels, train_labels.size())
print(train_data.mean(dim=0))
torch.save((train_data, train_labels), './train.pth')
print(test_data, test_data.size())
print(test_labels, test_labels.size())
print(test_data.mean(dim=0))
torch.save((test_data, test_labels),  './test.pth')
```
